```python
# This note book creates 6 datasets for Movies in US reqion with directors, actors, actresses, writers 
#and its ratings and votes.

# The dataset will be created in ./data/
# us_movies_ratings,csv
# us_movies_ratings_actor.csv
# us_movies_ratings_actress.csv
# us_movies_ratings_writer.csv
# us_movies_ratings_director.csv
# us_movies_ratings_self.csv

import pandas as pd
import time

#Load Datasets
start_time=time.ctime()
display(start_time)

## load the title basics dataset
title_akas_df = pd.read_csv('./data/title.akas.tsv', index_col=None, na_values=['NA'],sep='\t',low_memory=False)
```

```python
## load the title basics dataset
title_basics_df = pd.read_csv('./data/title.basics.tsv', index_col=None, na_values=['NA'],
sep='\t',low_memory=False)
```

```python
#load ratings dataset
title_ratings_df = pd.read_csv('./data/title.ratings.tsv', index_col=None, na_values=['NA'],
sep='\t',low_memory=False)
```

```python
# load the title crew dataset
#title_crew_df = pd.read_csv('./data/title.crew.tsv', index_col=None, na_values=['NA'],
#sep='\t',low_memory=False)
```

```python
# Filter data for region US
title_akas_us=title_akas_df[title_akas_df.region=='US']
```

```python
# Fiter only movies
title_basics_df_movie=title_basics_df.groupby('titleType').get_group('movie').copy()
#Filter only non adult movies
title_basics_df_movie= title_basics_df_movie[title_basics_df_movie.isAdult=='0'].copy()
```

```python
# Clean Filter data 
#remove null value|
#title_crew_df = title_crew_df[(title_crew_df.directors!='\\N') &(title_crew_df.writers!='\\N')].copy()
```

```python
#  rename columns to merge
title_akas_us.rename(columns={"titleId": "tconst"},inplace=True)
display(title_akas_us.head(5))
display(title_basics_df_movie.head(5))
```

```python
#result dataset for non adult movies only for region US
df_us_movies_non_adult_result = pd.merge(title_akas_us,
                 title_basics_df_movie,
                 on='tconst', 
                 how='inner')
```

```python
del title_basics_df_movie
del title_akas_us
```

```python
title_principals_df = pd.read_csv('./data/title.principals.tsv', index_col=None, na_values=['NA'],
sep='\t',low_memory=False)
```

```python
# load the names dataset
name_df = pd.read_csv('./data/name.basics.tsv', index_col=None, na_values=['NA'],
sep='\t',low_memory=False)
```

```python
#result dataset for non adult movies only for region US
principal_names_result_df = pd.merge(title_principals_df,
                 name_df,
                 on='nconst', 
                 how='inner')
```

```python
principal_names_result_df.head(10)
```

```python
del name_df
del title_principals_df
```

```python

principal_names_result_df = principal_names_result_df[['primaryName','ordering','birthYear','deathYear','tconst','category','job','characters']].copy()
```

```python
# Create Dataset for different Categories
principal_names_self_result_df = principal_names_result_df[principal_names_result_df.category=='self'].copy()
principal_names_director_result_df = principal_names_result_df[principal_names_result_df.category=='director'].copy()
principal_names_actor_result_df = principal_names_result_df[principal_names_result_df.category=='actor'].copy()
principal_names_actress_result_df = principal_names_result_df[principal_names_result_df.category=='actress'].copy()
principal_names_writer_result_df = principal_names_result_df[principal_names_result_df.category=='writer'].copy()
display(principal_names_director_result_df.head(5))

display(principal_names_self_result_df.head(5))

display(principal_names_actor_result_df.head(5))

display(principal_names_actress_result_df.head(5))

display(principal_names_writer_result_df.head(5))
```

```python
del principal_names_result_df
```

```python
# Dataset cleaning 

principal_names_director_result_df.rename(columns={"primaryName": "directorName","birthYear": "directorbirthYear","deathYear": "directordeathYear",},inplace=True)
principal_names_director_result_df = principal_names_director_result_df[['directorName','directorbirthYear','directordeathYear','tconst','ordering']].copy()
principal_names_self_result_df = principal_names_self_result_df[['primaryName','birthYear','deathYear','tconst','ordering']].copy()
principal_names_actor_result_df.rename(columns={"primaryName": "actorName","birthYear": "actorbirthYear","deathYear": "actordeathYear",},inplace=True)
principal_names_actor_result_df = principal_names_actor_result_df[['actorName','actorbirthYear','actordeathYear','tconst','ordering']].copy()
principal_names_actress_result_df.rename(columns={"primaryName": "actoressName","birthYear": "actoressbirthYear","deathYear": "actoressdeathYear",},inplace=True)
principal_names_actress_result_df=principal_names_actress_result_df[['actoressName','actoressbirthYear','actoressdeathYear','tconst','ordering']].copy()
principal_names_writer_result_df.rename(columns={"primaryName": "writerName","birthYear": "writerbirthYear","deathYear": "writerdeathYear",},inplace=True)
principal_names_writer_result_df=principal_names_writer_result_df[['writerName','writerbirthYear','writerdeathYear','tconst','ordering']].copy()
```

```python
display(principal_names_director_result_df.head(5))

display(principal_names_self_result_df.head(5))

display(principal_names_actor_result_df.head(5))

display(principal_names_actress_result_df.head(5))

display(principal_names_writer_result_df.head(5))
```

```python
actress_all_df.to_csv('./data/us_movies_ratings_actress.csv',sep=',',index=False)
del actress_all_df
```

```python
writer_all_df = pd.merge(df_us_movies_non_adult_ratings_result,
                 principal_names_writer_result_df,
                 on='tconst', 
                 how='inner')
writer_all_df=writer_all_df.loc[writer_all_df.groupby('tconst').ordering.idxmin()].copy()
writer_all_df.drop(columns=['ordering'],inplace=True)
```

```python
writer_all_df.to_csv('./data/us_movies_ratings_writer.csv',sep=',',index=False)
del writer_all_df
```

```python
self_all_df = pd.merge(df_us_movies_non_adult_ratings_result,
                 principal_names_self_result_df,
                 on='tconst', 
                 how='inner')
self_all_df=self_all_df.loc[self_all_df.groupby('tconst').ordering.idxmin()].copy()
self_all_df.drop(columns=['ordering'],inplace=True)
```

```python
self_all_df.to_csv('./data/us_movies_ratings_self.csv',sep=',',index=False)
del self_all_df
```
