```python
import pandas as pd

#Load Datasets

## load the title basics dataset
title_akas_df = pd.read_csv('./data/title.akas.tsv', index_col=None, na_values=['NA'],sep='\t',low_memory=False)


## load the title basics dataset
title_basics_df = pd.read_csv('./data/title.basics.tsv', index_col=None, na_values=['NA'],
sep='\t',low_memory=False)

#load ratings dataset
title_ratings_df = pd.read_csv('./data/title.ratings.tsv', index_col=None, na_values=['NA'],
sep='\t',low_memory=False)
```

```python
# Filter data for region US
title_akas_us=title_akas_df[title_akas_df.region=='US']
```

```python
# Fiter only movies
title_basics_df_movie=title_basics_df.groupby('titleType').get_group('movie').copy()
#Filter only non adult movies
title_basics_df_movie= title_basics_df_movie[title_basics_df_movie.isAdult=='0'].copy()
```

```python
#  rename columns to merge
title_akas_us.rename(columns={"titleId": "tconst"},inplace=True)
display(title_akas_us.head(5))
display(title_basics_df_movie.head(5))
```

```python
#result dataset for non adult movies only for region US
df_us_movies_non_adult_result = pd.merge(title_akas_us,
                 title_basics_df_movie,
                 on='tconst', 
                 how='inner')
```
